```python
#Filling null values with 0
raw = raw.replace(to_replace = np.nan, value = 0) 
transformed = transformed.replace(to_replace = np.nan, value = 0) 
```

```python
#Joining the two datasets
pop = raw.population
gdp = raw.gdp_per_capita
transformed["population"] = pop
transformed["gdp_cap"] = gdp

transformed["DATE"] = pd.to_datetime(transformed["DATE"], format = '%Y-%m-%d')


transformed = transformed[['CODE', 'COUNTRY', 'DATE', 'population', 'gdp_cap', 'HDI', 'TC', 'TD', 'STI']]
transformed = transformed.rename(columns={'CODE':'Code', 'COUNTRY':'Country', 'year':'Year', 'month':'Month', 'DATE':'Date', 'population':'Population', 'gdp_cap':'GDP_Cap', 'HDI':'HDI', 'TC':'Cases', 'TD':'Deaths', 'STI':'Stringency_Index'})
```

```python
#New dataset last 15
transformed.tail(15)
```

```python
#Pulling out datasets to work with 

transformed = transformed[transformed.Country != 'Kosovo']

country = transformed.Country.unique().tolist()
country_code = transformed.Code.unique().tolist()
world = transformed.Population.unique().tolist()

hdi_world = []
gdp_world = []
cases_country = []
death_country = []
stringency_index = []

for i in country:
    hdi_world.append((transformed.loc[transformed.Country == i, 'HDI']).sum()/294)
    gdp_world.append(transformed.loc[transformed.Country == i, 'GDP_Cap'].sum()/294)
    stringency_index.append(transformed.loc[transformed.Country == i, 'Stringency_Index'].sum()/294)
    cases_country.append(transformed.loc[(transformed["Country"] == i), "Cases"].sum())
    death_country.append(transformed.loc[(transformed["Country"] == i), "Deaths"].sum())
```

```python
#Importing library
!pip install pycountry_convert
import pycountry_convert as pc
```

```python
#Converting ISO codes
alpha2_code = []
for j in country_code:
    alpha2_code.append(pc.country_alpha3_to_country_alpha2(j))
continent_code = []
for j in alpha2_code:
    try:
        continent_code.append(pc.country_alpha2_to_continent_code(j))
    except:
        continent_code.append('Nameless') 

arg = pd.DataFrame(list(zip(country_code, country, world, cases_country, death_country, hdi_world, gdp_world, stringency_index, continent_code)), columns =['Code', 'Country', 'Population', 'Cases', 'Deaths', 'HDI', 'GDP_Cap','Stringency_Index', 'Continent']) 
arg = arg.replace({'AF':'Africa', 'AN':'Antarctica', 'AS':'Asia', 'EU':'Europe', 'NA':'North America', 'OC':'Oceania', 'SA':'South America'})
arg = arg.round(2)
```

```python
#First 7
arg.head(7)
```

**Visulaization**

```python
fig = px.scatter(arg, x="Stringency_Index", y="Deaths", size="Population", hover_name="Country", color='Continent', template='simple_white', size_max=50)
fig.update_layout(
    height=500,
    title_text="Stringency Index vs Deaths"
)
fig.show()
```

```python
fig = px.scatter(arg, x='Deaths', y='GDP_Cap', hover_name='Country', color='Continent', size='Population', template="simple_white", size_max=50)
fig.update_traces(textposition='top center')
fig.update_layout(
    height=500,
    title_text='Deaths vs GDP per Capita'
)
fig.show()
```
